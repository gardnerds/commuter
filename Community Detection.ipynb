```python
## use igraph and louvain for weighted directed modularity 
## all commuters

import igraph as ig
import louvain
import pandas as pd

years = [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]

for year in years:
    G = ig.Graph.Read_Ncol('bay_outer_tract_'+str(year)+'.txt', names=True, weights=True, directed=True)
    partition = louvain.find_partition(G, louvain.ModularityVertexPartition, weights="weight")
    print year, len(partition), "communities"
    print "total CC: ", G.transitivity_undirected()
    print "modularity",G.modularity(partition, weights="weight")

    commDict = {}
    partitionDict = {}
    for i in range(len(partition)):
        for j in range(len(partition[i])):
            vertex_id = partition[i][j]
            commDict["0"+G.vs[vertex_id]["name"]] = i
        partitionDict[i] = j
        print "community",i,"size:",j
        print "clustering coefficient barrat: ", np.asarray(G.transitivity_local_undirected(vertices=partition[i],weights='weight')).mean()
        print "clustering coefficient: ", np.asarray(G.transitivity_local_undirected(vertices=partition[i])).mean()
                

    print partitionDict
    print len(commDict)
    df = pd.DataFrame.from_dict(commDict, orient='index')
    print len(df)
    df.to_csv('weighted_directed_community_assignments_'+str(year)+'.csv')
```

```python
# poor agg stats

years = [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
label = 'low_income'

for year in years:
    
    print year
    
    comms_df = pd.read_csv('weighted_directed_community_assignments_poor_'+str(year)+'.csv')#, names=['w_trct','comm'], header=None)
    comms_df.columns = ['w_trct','w_comm']
    poor_df = pd.read_csv('bay_outer_tract_'+str(year)+'_poor.csv')

    df = pd.merge(poor_df, comms_df, on=['w_trct'])
    comms_df.columns = ['h_trct','h_comm']
    df = pd.merge(df, comms_df, on=['h_trct'])

    df['tot_dist'] = df[label]*df['dist']
    print "average commute distance (miles): ", df['tot_dist'].sum()/df[label].sum() 
    print "total commuters: ", df[label].sum() 
    df_agg = df.groupby(['w_comm', 'h_comm']).sum().reset_index()
    df_agg['avg_dist'] = df_agg['dist']/df_agg[label]

    df_agg = df_agg[['w_comm','h_comm',label,'avg_dist']]

    print len(df_agg)
    print df_agg
```
